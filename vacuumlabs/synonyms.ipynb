```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
print(nx.__version__)
print(np.__version__)
```

```python
G = nx.Graph()
G.add_edge('magic', 'watch')
G.add_edge('underdog', 'earth')
G.add_edge('earth', 'cake')
G.add_edge('uniform', 'balance')
G.add_edge('balance', 'ability')
G.add_edge('uniform', 'uniform')
G.add_edge('manager', 'watch')
G.add_edge('manager', 'manager')
G.add_edge('fake', 'earth')
G.add_edge('balance', 'cake')
G.add_edge('ability', 'uniform')
G.add_edge('underdog', 'magic')
```

```python
nx.draw_networkx(G)
plt.show()
```
